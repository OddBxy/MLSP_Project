## Prepare data

```python
def create_json_dataset(dataset_dir):
  DATASET_DIR = dataset_dir
  OUTPUT_JSON = dataset_dir + "/pena_dataset.json"

  data = []

  for filename in os.listdir(DATASET_DIR):
      if filename.endswith(".wav"):
          wav_path = filename
          txt_path = wav_path.replace(".wav", ".txt")
          txt_path = os.path.join(DATASET_DIR, txt_path)

          with open(txt_path, "r", encoding="utf-8") as f:
              caption = f.read().strip()

          data.append({
              "wav": wav_path,
              "caption": caption
          })

  with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
      json.dump({"data": data}, f, indent=2, ensure_ascii=False)

  print(f"JSON dataset created at : {OUTPUT_JSON}")
```

Split the json pena_dataset into train val & test dataset and build a dataset_root file in the expected format to root to the datasets.

```python
import random


dataset_json = "/content/labelled_dataset/data/slices/pena_dataset.json"
datasetSplits_output_dir = "/content/labelled_dataset/data/slices"
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

random.seed(0)

# Load original dataset
with open(dataset_json, "r", encoding="utf-8") as f:
    data = json.load(f)["data"]

# Shuffle entries
random.shuffle(data)

# Compute split sizes
n = len(data)
n_train = int(n * train_ratio)
n_val = int(n * val_ratio)
n_test = n - n_train - n_val

train_data = data[:n_train]
val_data = data[n_train:n_train+n_val]
test_data = data[n_train+n_val:]

# Save splits
os.makedirs(datasetSplits_output_dir, exist_ok=True)

split_files = {
    "train": os.path.join(datasetSplits_output_dir, "train.json"),
    "val": os.path.join(datasetSplits_output_dir, "val.json"),
    "test": os.path.join(datasetSplits_output_dir, "test.json")
}

for split_name, split_data in zip(["train", "val", "test"], [train_data, val_data, test_data]):
    with open(split_files[split_name], "w", encoding="utf-8") as f:
        json.dump({"data": split_data}, f, indent=2, ensure_ascii=False)
    print(f"Saved {split_name} split: {len(split_data)} entries -> {split_files[split_name]}")

# Create dataset_root.json
dataset_root = {
    "audiocaps" : datasetSplits_output_dir,
    "metadata": {
        "path": {
            "audiocaps":{
              "train" : split_files["train"],
              "val" : split_files["val"],
              "test" : split_files["test"]
            }
        }
    }
}

dataset_root_file = os.path.join(datasetSplits_output_dir, "dataset_root.json")
with open(dataset_root_file, "w", encoding="utf-8") as f:
    json.dump(dataset_root, f, indent=2)
print(f"Created dataset_root.json -> {dataset_root_file}")
```

#### Fix dependency conflicts

```python
import os

script_path = "/content/AudioLDM-training-finetuning/audioldm_train/train/latent_diffusion.py"

# The code we need to inject at the top
safety_code = """
import torch
import numpy
try:
    torch.serialization.add_safe_globals([numpy.core.multiarray.scalar, dict])
except:
    pass
"""

with open(script_path, 'r') as f:
    lines = f.readlines()

# Insert the code at line 2 (right after the first line)
lines.insert(1, safety_code + "\n")

with open(script_path, 'w') as f:
    f.writelines(lines)
```

```python
factory_path = "/content/AudioLDM-training-finetuning/audioldm_train/modules/clap/open_clip/factory.py"

with open(factory_path, 'r') as f:
    content = f.read()

# Replace the strict load with a permissive one
# This finds the specific line: torch.load(checkpoint_path, map_location=map_location)
# and adds weights_only=False
new_content = content.replace(
    "torch.load(checkpoint_path, map_location=map_location)",
    "torch.load(checkpoint_path, map_location=map_location, weights_only=False)"
)

with open(factory_path, 'w') as f:
    f.write(new_content)
```

```python
factory_path = "/content/AudioLDM-training-finetuning/audioldm_train/modules/clap/open_clip/factory.py"

with open(factory_path, 'r') as f:
    content = f.read()

# This handles the specific line 153 and any others
new_content = content.replace(".load_state_dict(ckpt)", ".load_state_dict(ckpt, strict=False)")
new_content = new_content.replace(".load_state_dict(state_dict)", ".load_state_dict(state_dict, strict=False)")

with open(factory_path, 'w') as f:
    f.write(new_content)
```
