```python
import os
import librosa
import numpy as np
from panns_inference import AudioTagging, labels
import torch

DOSSIER_AUDIO = './data/slices'

# Créer l'instance du tagger
# checkpoint_path=None va automatiquement télécharger le modèle Cnn14 pré‑entraîné
tagger = AudioTagging(checkpoint_path=None, device='cuda' if torch.cuda.is_available() else 'cpu')

for nom_fichier in os.listdir(DOSSIER_AUDIO):
    if nom_fichier.endswith('.wav'):
        chemin_audio = os.path.join(DOSSIER_AUDIO, nom_fichier)
        nom_base = os.path.splitext(nom_fichier)[0]

        print(f"Traitement de : {nom_fichier}...")

        try:
            # Charger l'audio avec librosa (mono, 32000 Hz recommandé)
            audio, sr = librosa.load(chemin_audio, sr=32000, mono=True)
            audio = audio[np.newaxis, :]  # forme (1, samples)

            # Inference audio tagging
            (clipwise_output, embedding) = tagger.inference(audio)

            # clipwise_output donne des scores pour les 527 classes AudioSet
            # labels.labels est la liste des labels associée
            tags_trouves = []
            seuil = 0.3

            for i, score in enumerate(clipwise_output[0]):
                if score >= seuil:
                    tags_trouves.append(labels[i])

            # Écrire dans un fichier TXT
            chemin_txt = os.path.join(DOSSIER_AUDIO, f"{nom_base}.txt")
            with open(chemin_txt, 'w', encoding='utf-8') as f:
                f.write(" ".join(tags_trouves))

            print(f"Tags enregistrés dans : {nom_base}.txt")

        except Exception as e:
            print(f"Erreur avec {nom_fichier} : {e}")

print("\nTraitement terminé.")
```
